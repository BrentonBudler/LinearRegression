Let us create the hypothesis function: $h_o(x) = \theta_0 + \theta_1x$

Partial Derivative of cost function wrt to $\theta_0$ is: <br>
$\frac{1}{n} \sum (y_{pred}-y_{actual})$

Partial Derivative of cost function wrt to $\theta_1$ is: <br>
$\frac{1}{n} \sum (y_{pred}-y_{actual}) x_i $

```python
n = len(X)
```

```python
def h(X,w):
    return (w[0]+ w[1]*X)
```

```python
def cost(w,X,y):
    return (0.5*n)*np.sum(np.square(h(X,w)-np.array(y)))
```

```python
def grad(w,X,y):
    g = [0]*2
    g[0] = (1/n)*np.sum(h(X,w)-np.array(y))
    g[1] = (1/n)*np.sum((h(X,w)-np.array(y))*X)
    return g
```

```python
def descent(w_new,w_prev, lr):
    print(w_prev)
    print(cost(w_prev,X,y))
    j=0 
    
    while True:
        w_prev = w_new
        w0 = w_prev[0] - lr*grad(w_prev,X,y)[0]
        w1 = w_prev[1] - lr*grad(w_prev,X,y)[1]
        w_new = [w0,w1]
        print(w_new)
        print(cost(w_new,X,y))
        
        if j>500:
            return w_new
        
        j+=1
        
```

```python
w = [0,-1]
```

```python
w = descent(w,w,.1)
print(w)
```
